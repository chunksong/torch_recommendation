```python
# import module for data processing
import pandas as pd
import time

ratings_df = pd.read_csv('./dataset/ml-latest-small/ratings.csv')
tags_df = pd.read_csv('./dataset/ml-latest-small/tags.csv')
movies_df = pd.read_csv('./dataset/ml-latest-small/movies.csv')
links_df = pd.read_csv('./dataset/ml-latest-small/links.csv')
```

```python
print('Table of movies haves columns:',', '.join(movies_df.keys()))
print('Table of rating haves columns:',', '.join(ratings_df.keys()))
print('Table of tags haves columns:',', '.join(tags_df.keys()))
print('Table of links haves columns:',', '.join(links_df.keys()))
```

```python
userIds = ratings_df.userId.unique()
movieIds = ratings_df.movieId.unique()

print("Unique user cnt: ", len(userIds), ", Unique movie cnt: ", len(movies_df))
print("Movie ids that miss a rating cnt: ", len(movies_df) - len(movieIds))
print("Here are movie ids that miss a rating: ", ', '.join([str(x) for x in set(movies_df.movieId)-set(movieIds)]))

# Building rating matrix
s_tstp = time.time()
ratings_mat = pd.DataFrame(index=userIds, columns=movieIds)
print("Building rating matrix...", end="")
for i in range(len(ratings_df)):
    ratings_mat.loc[ratings_df.iloc[i].userId][ratings_df.iloc[i].movieId] = ratings_df.iloc[i].rating
print("Done!-------- %d secs" %(time.time() - s_tstp))
ratings_mat.describe()


# TODO: Use a method with groupby op on ratings_df
# TODO: train test split
```
